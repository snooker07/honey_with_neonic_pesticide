# Clean

```python
#Copy the tables
df_first_clean = df_first.copy()
df_second_clean = df_second.copy()
```

### Define
Merge the two tables into one with the append function

### Code

```python
#merge the tables
df_clean = df_first_clean.append(df_second_clean, ignore_index = True)
```

### Test

```python
df_clean
```

```python
df_clean.info()
```

### Define
Correct the measurement unit inconsistency by converting kg units into lbs.
- 1 pound = 0.45359237 kg
- 1 kg = 2.20462262 pounds
- kg variables : nCLOTHIANIDIN, nIMIDACLOPRID, nTHIAMETHOXAM, nACETAMIPRID, nTHIACLOPRID

### Code

```python
# Create a lbl to kg converter
def kg_to_lbs(kg_variable):
    return kg_variable * 0.45359237
```

```python
# collect kg variable column names
kg_column_names = ['nCLOTHIANIDIN', 'nIMIDACLOPRID', 'nTHIAMETHOXAM', 'nACETAMIPRID', 'nTHIACLOPRID']
```

```python
# apply the converter function on the dataframe
for i in range(len(kg_column_names)):
    column = kg_column_names[i]
    df_clean[column] = df_clean[column].apply(lambda x: kg_to_lbs(x))
```

```python
# sum again the nCLOTHIANIDIN, nIMIDACLOPRID, nTHIAMETHOXAM, nACETAMIPRID, nTHIACLOPRID values in the nAllNeonic column
df_clean["nAllNeonic"] = df_clean[kg_column_names].sum(axis = 1)
```
